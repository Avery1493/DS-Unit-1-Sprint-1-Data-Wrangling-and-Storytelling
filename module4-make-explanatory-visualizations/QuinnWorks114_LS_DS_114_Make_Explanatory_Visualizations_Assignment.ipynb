```python
fig, ax = plt.subplots()
data=pd.Series([2.9,5.2,6.6,10.4,11.4,12,15.4,16.2,26.1,45.6])
reasons = ['Do not want to share the covers', 'Not enough space',
            'No longer physically intimate','Differnt temperature preferences',
            'Children', 'Frequent bathroom trips in the night',
            'Argument of fight', 'Night working/differnt sleeping times',
            'Sickness', 'Snoring']

  
data.plot.barh(width=.95)
ax.set_yticklabels(labels=reasons, color='grey')
#TITLE
ax.set_title('Why Do You Sleep Apart?', x=-.37, y=1.42, fontweight='bold',fontsize=20)
#SUBTITLE
ax.text(s='Respondents were asked to check all that apply', x=-36, y=13.23, 
        fontsize=17.5,
        color='black')
ax.yaxis.grid(b=None)
ax.get_xaxis().set_visible(False)

#https://stackoverflow.com/questions/30228069/how-to-display-the-value-of-the-bar-on-each-bar-with-pyplot-barh
for i, v in enumerate(data):
    ax.text(v + .5, i + -.2, str(v) +'%', color='grey')

#LINE 1
ax.text(s='_______________________________________________________________________________________________________',
        x=-36, y=12.7, color='grey')
#LINE 2
ax.text(s='_______________________________________________________________________________________________________',
        x=-36, y=10.7, color='grey')
#TEXT 3
ax.text(s=' SURVEY DATES                               NO. OF RESPONDENTS',
        x=-36, y=11.8, fontsize=10.8, fontweight='bold', color='grey')
#TEXT 4
ax.text(s=' July 9-10, 2014                     483',
        x=-36, y=11.1, fontsize=14, color='grey')


#########
display(example2)
```

```python
x = np.linspace(0, 6, 100)
y_1 = 5*x
y_2 = np.power(x, 2)
y_3 = np.exp(x/1.5)
plt.plot(x, y_1, color='cyan',linestyle='-.', label='$y=5*x$')
plt.plot(x, y_2, color='orange',linestyle='--',label='$y=x^2$')
plt.plot(x, y_3, color='pink',linestyle='-',label='$y=e^{3/2x}$')
plt.legend()
plt.show()
```
